```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Carrega o arquivo
load_file = "dados_compras.json"
purchase_file = pd.read_json(load_file, orient = "records")
purchase_file
```

## Informações Sobre os Consumidores

```python
# Implemente aqui sua solução
numeroTotalConsumidores = pd.DataFrame({'Número total de consumidores': [len(purchase_file.groupby('Login'))]})
numeroTotalConsumidores
```

## Análise Geral de Compras

```python
# Implemente aqui sua solução
purchase_file["Nome do Item"] = purchase_file["Nome do Item"].str.lower()
contItensExclusivos = len(purchase_file['Item ID'].unique())
mediaPrecoCompra = purchase_file["Valor"].mean()
numeroTotalCompras = len(purchase_file)
rendimentoTotal = purchase_file["Valor"].sum()
analiseGeralCompras = pd.DataFrame({ \
                                    'Número de itens exclusivos': [contItensExclusivos], \
                                    'Preço médio de compra': [mediaPrecoCompra], \
                                    'Número total de compras': [numeroTotalCompras], \
                                    'Rendimento total': [rendimentoTotal] \
                                   })
analiseGeralCompras
```

## Análise Demográfica

```python
# Implemente aqui sua solução
novoDataFrame = purchase_file
cont = novoDataFrame['Sexo'].value_counts()
percent = novoDataFrame['Sexo'].value_counts(normalize=True).mul(100).round(2)
analiseDemografica = pd.DataFrame({'Contagem': cont, '%': percent})
analiseDemografica.index.name = 'Gênero'
analiseDemografica
```

## Informações Demográficas Por Gênero

```python
# Implemente aqui sua solução
novoDataFrame = purchase_file
cont = novoDataFrame['Sexo'].value_counts()
percent = novoDataFrame['Sexo'].value_counts(normalize=True).mul(100).round(2)
analiseDemografica = pd.DataFrame({'Contagem': cont, '%': percent})
analiseDemografica.index.name = 'Gênero'
analiseDemografica
```

## Análise de Compras Por Gênero

```python
# Implemente aqui sua solução
numeroCompras = purchase_file['Sexo'].value_counts()
precoMedioCompra = purchase_file.groupby('Sexo')['Valor'].mean()
valorTotalCompra = purchase_file.groupby('Sexo')['Valor'].sum()
analiseComprasGenero = pd.DataFrame({'Quantidade de Compras': numeroCompras, 'Preço Médio de Compra': precoMedioCompra, 'Valor Total de Compra': valorTotalCompra})

baseDF = purchase_file
comprasFaixaEtaria = baseDF.groupby(pd.cut(baseDF["Idade"], np.arange(0, 61, 5))).count().rename(columns = {'Login': 'Quantidade de compras'}, inplace = False).drop(columns=['Sexo', 'Idade','Item ID', 'Valor', 'Nome do Item'])

analiseComprasGenero
# comprasFaixaEtaria
```

## Consumidores Mais Populares (Top 5)

```python
# Implemente aqui sua solução
baseDF = purchase_file.groupby(['Login', 'Idade']).count().reset_index().sort_values("Valor", ascending=False).head(5)
numeroCompras = purchase_file['Login'].value_counts()
precoMedioCompra = purchase_file.groupby('Login').mean()['Valor']
valorTotalCompra = purchase_file.groupby('Login').sum()['Valor']

consumidoresMaisPopulares = pd.DataFrame({'Número de compras': numeroCompras, 'Preço Médio de Compra': precoMedioCompra, 'Valor Total de Compra': valorTotalCompra})
consumidoresMaisPopulares = consumidoresMaisPopulares[consumidoresMaisPopulares.index.isin(baseDF['Login'])]
consumidoresMaisPopulares
```

## Itens Mais Populares

```python
# Implemente aqui sua solução
mostPopularItemsId = purchase_file['Item ID'].value_counts().sort_values(ascending=False).head(5)
mostPopularItemsDf = purchase_file[purchase_file['Item ID'].isin(mostPopularItemsId.index.values)]

numeroComprasPopulares = mostPopularItemsDf[['Item ID', 'Nome do Item']].value_counts()
precoItemsPopulares = mostPopularItemsDf.drop_duplicates(['Item ID', 'Nome do Item', 'Valor']).reset_index().set_index(['Item ID', 'Nome do Item'])['Valor']
valorTotalCompraPopulares = mostPopularItemsDf.groupby(['Item ID', 'Nome do Item'])['Valor'].sum()

itensMaisPopulares = pd.DataFrame({'Número de Compras': numeroComprasPopulares, 'Preço do Item': precoItemsPopulares,'Valor Total de Compra': valorTotalCompraPopulares})
itensMaisPopulares
```

## Itens Mais Lucrativos

```python
# Implemente aqui sua solução
mostProfitableItemsId = purchase_file.groupby('Item ID').sum()['Valor'].sort_values(ascending=False).head(5)
mostProfitableItemsDf = purchase_file[purchase_file['Item ID'].isin(mostProfitableItemsId.index.values)]

numeroComprasLucrativos = mostProfitableItemsDf[['Item ID', 'Nome do Item']].value_counts()
valorTotalCompraLucrativos = mostProfitableItemsDf.groupby(['Item ID', 'Nome do Item'])['Valor'].sum()
precoItemsLucrativos = mostProfitableItemsDf.drop_duplicates(['Item ID', 'Nome do Item', 'Valor']).reset_index().set_index(['Item ID', 'Nome do Item'])['Valor']

itensMaisLucrativos = pd.DataFrame({'Número de Compras': numeroComprasLucrativos, 'Preço do Item': precoItemsLucrativos,'Valor Total de Compra': valorTotalCompraLucrativos})
itensMaisLucrativos
```
